```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use('ggplot')
```

```python
#QS=QUICK SURVEY--IMPORT CSV FILE
QS= pd.read_csv('Quick Survey - DATA TABLE.csv')
```

```python
FR=pd.read_csv('Quick Survey-Form responses 1.csv')
```

```python
   #DATA BREAKDOWN     
QS1= QS['AGE']
QS2= QS['MOST PREFERRED FLAVOR']
QS3= QS['NO OF PEOPLE PER FLAVOR']
QS4= QS['LEAST PREFERRED FLAVOR']
QS5= QS['NO OF PEOPLE PER FLAVOR.1']
QS6= QS['PURCHASING FACTOR']   #price, flavor
QS7= QS['PURCHASING FACTOR [BRAND]']   #brand
QS8= QS['SEX']
QS9= QS['`NO OF PEOPLE']   #sex
QS10= QS['PRICE FACTOR[ EXPENSIVE]']
QS11= QS['NO OF CHOICES']
QS12= QS['CHANGE IN FAVORITE FLAVOR IN THE LAST SIX  MONTH']
QS13= QS['CHANGE CHOICES']
QS14 =QS['NO OF PEOPLE PER PURCHASING FACTOR.1']  #BRAND CHOICES
QS15 =QS['NO OF PEOPLE PER PURCHASING FACTOR']
```

```python
#FACTORS CUSTOMERS CONSIDER WHEN MAKING ICE-CREAM PURCHASES 
#IN TERMS OF PRICE, BRAND POPULARITY AND FLAVOR.
#EXPENSIVE PRICE
QS10=QS10.head(3)        #split
QS11=QS11.head(3)        #split
#convert NO OF CHOICES TO INT
QS11=QS11.astype('int')
#BARCHAT SHOWING FACTORS CONSIDER BASED ON PRICE OF ICE-CREAM FLAVOR
#LABELS
plt.title('PURCHASE OF ICECREAM BASED ON EXPENSIVE PRICE', fontsize=12,fontfamily='Verdana') 
plt.ylabel('NO OF CHOICES',fontsize=10,fontfamily='Verdana')   
plt.xlabel('CHOICES',fontsize=10,fontfamily='Verdana')
cols=('#FF7F57','#DC143C','#DA70D6')

#bar plot
plt.bar(QS10,QS11,color=cols,width=0.7)
sns.set_style('ticks')
print('The barchart below shows that majority of the sample poplulation would still buy their favorite icecream flavor even when it is more expensive than normal.')
```

```python
#BRAND
QS7=QS7.head(2)   #split
QS14=QS14.head(2)  #split

#CONVERT NO OF CHOICES TO INT
QS14=QS14.astype('int')

#labels
plt.title('FACTOR CONSIDERED BASED ON THE BRAND OF ICE-CREAM',fontsize=12,fontfamily='Verdana')
plt.ylabel('NO OF CHOICES',fontsize=12,fontfamily='Verdana')
plt.xlabel('BRAND FACTOR',fontsize=12,fontfamily='Verdana')
cols=('#04D8B2','#00FFFF')

#plot
plt.bar(QS7,QS14,color=cols,linewidth=5)
sns.set_style('darkgrid')
print('The chart below shows that according to the choices of the sample population, most people prefer to purchase icecream from the brand that makes their favorite icecream better')
```

```python
#FLAVOR VERSUS PRICE
QS6=QS6.head(3)    #split
QS15=QS15.head(3)  #split

#convert no of choices to int

QS15=QS15.astype('int')

#title
plt.title('FACTOR CONSIDERED BASED ON PREFERRED FLAVOR AND PRICE',fontsize=12,fontfamily='Verdana')
plt.ylabel('NO OF CHOICES',fontsize=12,fontfamily='Verdana')
plt.xlabel('PREFERRED FLAVOR AND PRICE FACTOR',fontsize=12,fontfamily='Verdana')
cols=('#C875C4','#FFC0CB','#FF81C0')
#plot
plt.bar(QS6,QS15,color=cols,linewidth=5)
sns.set_style('ticks')
sns.despine()
print('The barchart below shows that when making purchases of their icecream flavor,majority of the sample population would consider their favorite flavor over the price of the icecream')
```

```python
#Change in flavor preference
QS12=QS12.head(2)    #split
QS13=QS13.head(2)    #split

#convert no of choices to int
QS13=QS13.astype('int')

#LABEL
plt.title('CHANGE IN FLAVOR PREFERENCE IN THE LAST SIX MONTH',fontsize=13,fontfamily='Verdana')
plt.ylabel('NO OF CHOICES')
plt.xlabel('OPTIONS IN TERMS OF (YES/NO)')
cols=('#DDA0DD','#580F41')



#PLOT
plt.bar(QS12,QS13,color=cols,linewidth=7)
sns.set_style('ticks')
sns.despine()

print('The barchart shows that majority of the sample population has not changed their favorite icecream flavor in the last six month')
```

```python
#most preferred flavor
QS2=QS2.head(9)   #split
QS3=QS3.head(9)   #split

#LABEL
QS3=QS3
label=QS2
cols=('#FF6347','#FE420F','#DDA0DD','#C1F90A','#929591','#7E1E9C','#C20078','#0000FF')


#plot
plt.pie(QS3,labels=QS2,colors=cols,explode=(0.3,0,0,0,0,0,0,0,0),startangle=180,radius=3,shadow=True,autopct='%0.01f%%')
plt.legend(bbox_to_anchor=(2.0,1.1))
sns.set_style('ticks')
sns.despine()
print('The chart below shows that most people in the sample population prefer strawberry flavored ice-cream')
```
